## investigating the relationship between salaries and company size

#### the dataframe used named'jobmarket' is job market data for 500 employees, and importantly,salaries earned in diferently-sized companies.
#### according to the attributes of the dataset, the dependent variable in this case will be the salary_usd while the independent variable is the company_size
#### in this example,the linear regression plot indicates the best line of fit of the relationship between the two variables.
#### the independent variable attribute can be swithced back and forth based on the subject of investigation in the jobmarket.

```python



#importing the necessary libraries to perform linear regression analysis

import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn import datasets, linear_model
import seaborn as sb

# loading the dataset

jobmarket=pd.read_csv('jobmarket.csv')
print(jobmarket)

#cleaning the data from null values
cleaning=pd.read_csv('jobmarket.csv')
jobmarket=cleaning.dropna()
print(jobmarket)

#splitting the dataset into training and testing sets

trainingset,testingset=train_test_split(jobmarket,test_size=0.5,random_state=2)
print('training data is')
print(trainingset)
print('testing data is')
print(testingset)


#splitting the data into dependent and independent variables

training_independent_variable=trainingset.values[:, 2]
training_dependent_variable=trainingset.values[:, -3]
print('independent variables for the training data is:')
print(training_independent_variable)
print('dependent variables for the training data is:')
print(training_dependent_variable)

#visualize the relationship between the variables in training data set using a scatter graph

plt.scatter(training_independent_variable,training_dependent_variable,color='Blue')
plt.title('data for training dataset')
plt.xlabel('company size')
plt.ylabel('salary')
plt.show()
```
